```python
import pandas as pd
df=pd.read_csv("/content/superstore.csv")
df.head()
```

```python
# Convert 'Order.Date' and 'Ship.Date' columns to datetime format
df['Order.Date'] = pd.to_datetime(df['Order.Date'])
df['Ship.Date'] = pd.to_datetime(df['Ship.Date'])

# Verify the conversion
print(df[['Order.Date', 'Ship.Date']].head())
```

```python
# Create 'Month-Year' column to group data by month and year
df['Month-Year'] = df['Order.Date'].dt.to_period('M')

# Verify the new column
print(df[['Order.Date', 'Month-Year']].head())
```

```python
# Check for missing values
print(df.isnull().sum())

# Fill or drop missing values (example: filling with 0 for numerical columns)
df.fillna({'Sales': 0, 'Shipping.Cost': 0, 'Discount': 0}, inplace=True)

# Check again for missing values
print(df.isnull().sum())
```

```python
# Assuming profit = sales - shipping cost - discount impact on sales
df['Profit'] = df['Sales'] - df['Shipping.Cost'] - (df['Discount'] * df['Sales'])
```

```python
# Example: Filtering by region (e.g., 'California')
california_sales = df[df['State'] == 'California']

# Example: Grouping by 'Month-Year' and calculating total sales and profit
monthly_sales = df.groupby('Month-Year')[['Sales', 'Profit']].sum().reset_index()

print(monthly_sales.head())
```

```python
# Save the cleaned data to a new CSV file
df.to_csv('Cleaned_Superstore_Sales.csv', index=False)
```
